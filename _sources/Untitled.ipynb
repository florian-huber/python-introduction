```python
import pandas as pd
```

```python
data1 = pd.DataFrame({"K": [1, 2, 3, 4],
                     "A": 4 * ["a"],
                     "B": 4 * ["b"]})
data2 = pd.DataFrame({"K": [3, 4, 5, 6],
                     "B": ["b"] + 3 * ["b'"],
                     "C": 4 * ["c"]})
data2
```

```python
data1
```

```python
pd.merge(data1, data2)
```

```python
pd.merge(data1, data2, on="K")
```

```python
data1.merge(data2)
```

```python
data1.merge(data2, on="K")
```

```python
data1.merge(data2, on="K", suffixes=["_data1", "_data2"])
```

```python
data1.merge(data2, on="K", how="inner")
```

```python
data1.merge(data2, on="K", how="outer") # right, left...
```

```python
data1.merge(data2, on=["K", "B"], how="outer")
```

```python
pd.concat([data1, data2])
```

```python
pd.concat([data1, data2], ignore_index=True)
```

```python
pd.concat([data1, data2], axis=1)
```

```python
pd.concat([data1, data2], axis=0)
```
